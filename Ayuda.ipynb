```python
import matplotlib.pyplot as plt
import pandas as pd
import os

def load_data():
    df = pd.read_csv("files/input/shipping-data.csv")
    return df

# Configuración de rutas
carpeta_raiz_docs = 'docs'
if not os.path.exists(carpeta_raiz_docs):
    os.makedirs(carpeta_raiz_docs)
ruta_guardado_shipping_per_warehouse = os.path.join(carpeta_raiz_docs, "shipping_per_warehouse.png")
ruta_guardado_Mode_of_Shipment = os.path.join(carpeta_raiz_docs, "mode_of_shipment.png")
ruta_guardado_average_customer_rating = os.path.join(carpeta_raiz_docs, "average_customer_rating.png")
ruta_guardado_weight_distribution = os.path.join(carpeta_raiz_docs, "weight_distribution.png")

# Gráfico de barras Shipping per warehouse
def create_visual_for_shipping_per_warehouse(df):
    df = df.copy()
    fig, ax = plt.subplots()
    counts = df.Warehouse_block.value_counts()
    counts.plot.bar(
        ax=ax,
        title="Shipping per Warehouse",
        xlabel="Warehouse block",
        ylabel="Record Count",
        color="tab:blue",
        fontsize=8
    )
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    fig.savefig(ruta_guardado_shipping_per_warehouse)
    plt.close(fig)

# Gráfico Mode of Shipment
def create_visual_for_mode_of_shipment(df):
    df = df.copy()
    fig, ax = plt.subplots()
    counts = df.Mode_of_Shipment.value_counts()
    counts.plot.pie(
        ax=ax,
        title="Mode of shipment",
        wedgeprops=dict(width=0.35),
        ylabel="Record Count",
        color=["tab:blue", "tab:orange", "tab:green"],
    )
    fig.savefig(ruta_guardado_Mode_of_Shipment)
    plt.close(fig)

# Gráfico Average Customer Rating
def create_visual_for_average_customer_rating(df):
    df = df.copy()
    fig, ax = plt.subplots()
    df = (
        df[["Mode_of_Shipment", "Customer_rating"]]
        .groupby("Mode_of_Shipment")
        .describe()
    )
    df.columns = df.columns.droplevel()
    df = df[["mean", "min", "max"]]
    ax.barh(
        y=df.index.values,
        width=df["max"].values-1,
        left=df["min"].values,
        height=0.9,
        color="lightgray",
        alpha=0.8
    )
    colors = [
        "tab:green" if value >=3.0 else "tab:orange" for value in df["mean"].values
    ]
    ax.barh(
        y=df.index.values,
        width=df["mean"].values-1,
        left=df["min"].values,
        height=0.5,
        color=colors,
        alpha=1.0
    )
    ax.set_title("Average Customer Rating")
    ax.spines["left"].set_color("gray")
    ax.spines["bottom"].set_color("gray")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    fig.savefig(ruta_guardado_average_customer_rating)
    plt.close(fig)
    return colors

# Gráfico Weight Distribution
def create_visual_for_weight_distribution(df):
    df = df.copy()
    fig, ax = plt.subplots()
    df.Weight_in_gms.plot.hist(
        ax=ax,
        title="Shipped Weight Distribution",
        color="tab:orange",
        edgecolor="white"
    )
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    fig.savefig(ruta_guardado_weight_distribution)
    plt.close(fig)

# Ejecutar las funciones
df = load_data()
create_visual_for_average_customer_rating(df)
create_visual_for_mode_of_shipment(df)
create_visual_for_shipping_per_warehouse(df)
create_visual_for_weight_distribution(df)
```
